```python
import pandas as pd
```

```python
#Преобразования номера в вид 79991234567
#data - название входящего dataframe
#name - имя колонки содержащей номер
#new_name - имя новой колонки с номером
def number_norm(data, name, new_name):  
    
    
    #создаем dataframe для внутренних преобразований
    df = data.copy()
    
    #Вставляем новый столбец для номера и проверяем есть ли он уже
    try:
        df.insert(loc=0 , column=new_name, value=df[name])
    except ValueError:
        #если он уже есть проверяем что новый столбец соответствует старому
        if df[new_name][0] == df[name][0]:
            pass
        else:            
            print('Выбери другое имя нового столбца!')
            return
    
    #функция для преобразования строки с номером
    #n - строка
    def number_change(n):
        #посимвольно проверяем строку n, цифры записываем в строку r, остальное отбрасываем
        try:
            r = ''
            for i in n:
                if i.isdigit():
                    r += str(i)
        #в случае возникновения проблемы при проверке возвращаем None
        except TypeError:
            return None
        #при удачном преобразовании:
        else:
            # создаем новую переменную в которую будем записывать конечный результат, 
            # если получившийся набор цифр не соответствует нашим требованиям к номеру будем возвращать None
            r1 = None
            # проверяем что номер не более 11 цифр и удаляем лишнее
            if len(r) > 10:        
                r = r[0:11]
                # оставляем если в начяле стоит 7
                if r[0] == '7':
                    r1 = r
                # заменяем в начале 8 на 7
                elif r[0] == '8':
                    r1 = '7' + r[1:]
            # если номер записан без 8 или +7 добавляем 7
            elif len(r) == 10:
                r1 = '7' + r
        return r1
    
    display(df.shape)
    df[new_name] = df[new_name].map(number_change)
    df_dropna = df.dropna(subset=[new_name], ignore_index=True)
    display(df_dropna.info())
    display(df_dropna)
    return(df_dropna)
```

```python
piranya = pd.read_csv('пиранья.csv', dtype=str)
display(piranya.head(5))
piranya.info()
```

```python
piranya['Телефон'].isnull().value_counts()
```

```python
piranya_mod = number_norm(data = piranya, name = 'Телефон', new_name='number')
piranya_mod.to_excel('Пиранья обработанно.xlsx')
```
